```python
def update_velocity(particle, velocity, pbest, gbest, w_min=0.5, max=1.0, c=0.1):
  # Initialise new velocity array
  num_particle = len(particle)
  new_velocity = array([0.0 for i in range(num_particle)])
  # Randomly generate r1, r2 and inertia weight from normal distribution
  r1 = uniform(0,max)
  r2 = uniform(0,max)
  w = uniform(w_min,max)
  c1 = c
  c2 = c
  # Calculate new velocity
  for i in range(num_particle):
    new_velocity[i] = w*velocity[i] + c1*r1*(pbest[i]-particle[i])+c2*r2*(gbest[i]-particle[i])
  return new_velocity

def update_position(particle, velocity):
  # Move particles by adding velocity
    new_particle = particle + velocity
    return new_particle

# def pso_2d(population, dimension, position_min, position_max, generation, fitness_criterion, function):
def pso_2d(population, dimension, position_min, position_max, generation, function):
  # Initialisation
  # Population
  particles = array([[uniform(position_min, position_max) for j in range(dimension)] for i in range(population)])
  # Particle's best position
  pbest_position = particles
  # Fitness
  pbest_fitness = [function(p) for p in particles]
  # Index of the best particle
  gbest_index = np.argmin(pbest_fitness)
  # Global best particle position
  gbest_position = pbest_position[gbest_index]
  # Velocity (starting from 0 speed)
  velocity = zeros((population,dimension))
  # Loop for the number of generation
  vmax = 0.1*(position_max-position_min)
  vmin = -vmax

  pbestf_curve = []

  for t in range(generation):
    # Stop if the average fitness value reached a predefined success criterion
    # if np.average(pbest_fitness) <= fitness_criterion:
    #   break
    # else:
      # for n in range(population):
      #   # Update the velocity of each particle
      #   velocity[n] = update_velocity(particles[n], velocity[n], pbest_position[n], gbest_position)
      #   # Move the particles to new position
      #   particles[n] = update_position(particles[n], velocity[n])
    for n in range(population):
      # Update the velocity of each particle
      velocity[n] = update_velocity(particles[n], velocity[n], pbest_position[n], gbest_position)
      for i in range(dimension):
        velocity[n,i] = maximum(velocity[n,i],vmin)
        velocity[n,i] = minimum(velocity[n,i],vmax)
      # Move the particles to new position
      particles[n] = update_position(particles[n], velocity[n])
      for i in range(dimension):
        particles[n,i] = maximum(particles[n,i],vmin)
        particles[n,i] = minimum(particles[n,i],vmax)

    # Calculate the fitness value
    pbest_fitness = [function(p) for p in particles]
    # Find the index of the best particle
    gbest_index = np.argmin(pbest_fitness)
    # Update the position of the best particle
    gbest_position = pbest_position[gbest_index]
    pbestf_curve = np.append(pbestf_curve, np.average(pbest_fitness))

  # print('Global Best Position: ', gbest_position)
  # print('Best Fitness Value: ', min(pbest_fitness))
  # print('Average Particle Best Fitness Value: ', np.average(pbest_fitness))
  # print('Number of Generation: ', t)

  return gbest_position, pbest_fitness, t, pbestf_curve, particles

# population = 30
# dimension = 4
# position_min = 0
# position_max = 1
# generation = 1000

# gbest_position, pbest_fitness, t, pbestf_curve, p = pso_2d(population, dimension, position_min, position_max, generation, f10)

# print('Global Best Position: ', gbest_position)
# print('Best Fitness Value: ', np.amin(pbest_fitness))
# print('Average Particle Best Fitness Value: ', np.average(pbest_fitness))
# print('Number of Generation: ', t)


# plt.rcParams["figure.figsize"] = [7.50, 3.50]
# plt.rcParams["figure.autolayout"] = True

# plt.title("Objective space PSO")

# plt.plot(pbestf_curve, color="red")

# plt.xlabel("Iteration")
# plt.ylabel("Best score so far")

# plt.show()
```

```python
def Intensity(N,Xprey,X):
  
  di = np.empty(N)
  S = np.empty(N)

  for i in range(N-1):
    di[i] = (norm((X[i,:]-Xprey+np.spacing(1))))**2
    S[i] = (norm((X[i,:]-X[i+1,:]+np.spacing(1))))**2
    # di = np.append(di,di_i)
    # S = np.append(S,S_i)
  
  di[N-1] = (norm((X[N-1,:]-Xprey+np.spacing(1))))**2
  S[N-1] = (norm((X[N-1,:]-X[1,:]+np.spacing(1))))**2
  # di = np.append(di,di_N)
  # S = np.append(S,S_N)

  I = np.empty(N)
  for i in range(N):
      r2 = random_sample()
      I[i] = r2*S[i]/(4*np.pi*di[i])
      # I = np.append(I, I_i)
  return I

def initialization(SearchAgents, dimension, up, down):
  X=random_sample((SearchAgents,dimension))*(up-down)+down
  return X

def fun_calcobjfunc(func, X):
  K = np.size(X, axis=0)
  Y = np.empty(K)
  for i in range(K):
    Y[i] = func(X[i,:])
    # Y = np.append(Y,Yi)
  return Y

def HBA(objfunc, dimension, lowerbound, upperbound, Max_iterations, SearchAgents):
  
  beta = 6
  C = 2
  vec_flag = [1,-1]
  X = initialization(SearchAgents, dimension, upperbound, lowerbound)
  fitness = fun_calcobjfunc(objfunc, X)

  GYbest = fitness.min(axis=0)
  gbest = fitness.argmin(axis=0)
  Xprey = X[gbest,:]

  Xnew = np.empty([SearchAgents,dimension])
  CNVG = np.empty(Max_iterations)

  for t in range(0,Max_iterations):
    alpha=C*np.exp(-(t+1)/(Max_iterations))
    I=Intensity(SearchAgents,Xprey,X)
    for i in range(SearchAgents):

        r = random_sample()
        F = vec_flag[np.floor(2*random_sample()).astype(int)]

        for j in range(dimension):
            di = Xprey[j]-X[i,j]
            if r<0.5:
              r3 = random_sample()
              r4 = random_sample()
              r5 = random_sample()  
                
              Xnew[i,j] = Xprey[j] + F*beta*I[i]*Xprey[j] + F*r3*alpha*di*abs(np.cos(2*np.pi*r4)*(1-np.cos(2*np.pi*r5)))
            else :
              r7 = random_sample()
              Xnew[i,j] = Xprey[j]+F*r7*alpha*di

        FU=Xnew[i,:]>upperbound
        FL=Xnew[i,:]<lowerbound
        Xnew[i,:]=(Xnew[i,:]*(np.logical_not(FU+FL)))+upperbound*FU + lowerbound*FL
        tempFitness = objfunc(Xnew[i,:])
        if tempFitness<fitness[i]:
          fitness[i]=tempFitness
          X[i,:]= Xnew[i,:]
          
    FU = X>upperbound
    FL = X<lowerbound
    X = (X*np.logical_not(FU+FL)) + upperbound*FU + lowerbound*FL
    Ybest = fitness.min(axis=0)
    index = fitness.argmin(axis=0)
    CNVG[t] = amin(Ybest)
    if Ybest<GYbest:
        GYbest=Ybest
        Xprey = X[index,:]
  Food_Score = GYbest

  return Xprey, Food_Score, CNVG

# dim=30
# T=1000
# Lb=-30
# Ub=30
# N=30
# xmin, fmin, CNVG = HBA(f3, dim, Lb, Ub, T, N)

# print("x_pray = ", xmin)
# print("food score = ", fmin)

# plt.rcParams["figure.figsize"] = [7.50, 3.50]
# plt.rcParams["figure.autolayout"] = True

# plt.title("Objective space HBA")
# plt.plot(CNVG, color="green")

# plt.xlabel("Iteration")
# plt.ylabel("Best score so far")

# plt.show()
```

# DTBO

```python
def DTBO(SearchAgents,Max_iterations,lowerbound,upperbound,dimension,fitness):
  
  lowerb=ones((1,dimension))*lowerbound                          # Lower limit for variables
  upperb=ones((1,dimension))*upperbound                         # Upper limit for variables

  X = array([])

  for i in range(dimension):
    Xi = lowerb[:,i]+random_sample((SearchAgents,1))*(upperb[:,i] - lowerb[:,i])
    X = np.append(X,Xi)

  X = X.reshape(SearchAgents,dimension)
  fit=array([])

  for i in range(SearchAgents):
    L = X[i,:]
    fit_i = fitness(L)
    fit = np.append(fit,fit_i)
  best_so_far=np.array([])
  
  for t in range(1,Max_iterations+1):
    best, blocation = fit.min(axis=0),fit.argmin(axis=0)
    if t==1:
      Xbest = X[blocation,:]
      fbest = best
    elif best<fbest:
      fbest = best
      Xbest = X[blocation,:]

    XF = concatenate([X, fit[None,:].conj().transpose()], axis=1)
    XFsort = XF[XF[:,dimension].argsort(),]
    X = XFsort[:,0:dimension]
    fit=XFsort[:,dimension].conj().transpose()
    N_DI = around(0.2*SearchAgents*(1-t/Max_iterations)).astype(int)

    if N_DI == 0:
      DI=XFsort[0,0:dimension]
      F_DI=XFsort[0,dimension].conj().transpose()
    else :
      DI=XFsort[0:N_DI,0:dimension]
      F_DI=XFsort[0:N_DI,dimension].conj().transpose()

    for i in range(SearchAgents):

      # Fase 1 : training by the driving instructor

      # k_i = np.random.default_rng()
      if N_DI == 0:
        k_i = 0
        F_DI_ki=F_DI
      else:
        k_i = np.random.default_rng().choice(N_DI, 1)
        F_DI_ki=F_DI[k_i]

      DI_ki = DI[k_i]
      I=around(1+random_sample((1,1)))

      if F_DI_ki< fit[i]:
        X_P1= X[i,:]+random_sample((1,1)) * (DI_ki - I * X[i,:])
      else:
        X_P1= X[i,:]+random_sample((1,1)) * (X[i,:] - DI_ki)  
          
      X_P1 = maximum(X_P1,lowerb)
      X_P1 = minimum(X_P1,upperb)
      X_P1 = X_P1.flatten()
          
      F_P1 = fitness(X_P1)
      if F_P1 <= fit[i]:
        X[i,:] = X_P1
        fit[i]=F_P1

      # Akhir dari fase 1
      # Fase 2: learner driver patterning from instructor skills

      PIndex=0.01+0.9*(1-t/Max_iterations)
      X_P2 =  PIndex * X[i,:] + (1-PIndex) * DI_ki
      X_P2 = maximum(X_P2,lowerb)
      X_P2 = minimum(X_P2,upperb)
      X_P2 = X_P2.flatten()

      F_P2 = fitness(X_P2)
      if F_P2 <= fit[i]:
        X[i,:] = X_P2
        fit[i] = F_P2

      # Akhir dari fase 2
      # Fase 3: personal practice

      R=0.05
      X_P3= X[i,:] + (1-2*random_sample((1,dimension))) * R * (1-t/Max_iterations) * X[i,:]
      X_P3= maximum(X_P3,lowerb)
      X_P3 = minimum(X_P3,upperb)
      X_P3 = X_P3.flatten()
            
      F_P3 = fitness(X_P3)
      if F_P3 <= fit[i]:
        X[i,:] = X_P3
        fit[i] = F_P3

      # Akhir dari fase 3

    best_so_far=np.append(best_so_far,fbest)

  return fbest, Xbest , best_so_far

lowerbound = 0
upperbound = 1
SearchAgents = 4
Max_iterations = 1000
dimension = 30

Best_score,Best_pos,DTBO_curve=DTBO(SearchAgents,Max_iterations,lowerbound,upperbound,dimension,f10)

# print(Best_score)
# print(Best_pos)

# plt.rcParams["figure.figsize"] = [7.50, 3.50]
# plt.rcParams["figure.autolayout"] = True

# plt.title("Objective space DTBO")
# plt.plot(DTBO_curve, color="blue")

# plt.xlabel("Iteration")
# plt.ylabel("Best score so far")

# plt.show()
```
